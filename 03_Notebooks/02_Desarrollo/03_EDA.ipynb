## 3. EDA categoricas

Eliminar la variable de *date*, ya que se va a utilizar despues para graficar.

```python
incluir = cat.columns.to_list()
del incluir[0]
```

Función para ver las frecuencias de cada valor categoricos

```python
def frecuencias_cat(df_cat):
    df_cat = df_cat.astype(str)
    
    resultado = df_cat.apply(lambda x: x.value_counts(normalize=True)) \
                      .T.stack() \
                      .to_frame().reset_index() \
                      .rename(columns={'level_0': 'Variable', 'level_1': 'Valor', 0: "Frecuencia"}) \
                      .sort_values(by=['Variable', 'Frecuencia'])
    return resultado
```

Vemos el resultado

```python
pd.set_option('display.max_rows', None)

frecuencias_cat(cat[incluir])
```

Notar que:

1.- La mayoria de datos son durante la semana, pero estan repartidos de forma homogenea entre los 7 dias de la semana.

2.- El tipo de carga de sobre sale es la media.

Función para graficar las variables categoricas.

```python
def graficos_eda_categoricos(cat):
    
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    ax = ax.flat 

    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

```python
graficos_eda_categoricos(cat[incluir])
```

Vemos las mismas conclusiones que cuando revisamos la frecuencia de los valores categoricos.

### 4.1 Analisis estadistico

```python
def estadisticos_cont(num):
    estadisticos = num.describe().T
    estadisticos['median'] = num.median()
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    return(estadisticos)
```

## 5. Correlacion

```python
corr_matrix = num.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()
```

Comentario:

- Las emisiones de CO2 estan relacionadas fuertemente con el uso de voltaje y el factor de potencia con corriente retrasada. La conclusión directa, aunque no definitiva, es que a mayor uso de voltaje y/o potencia aumenta las emisiones de CO2.

- El uso de voltaje esta relacionado con el factor de potencia con corriente retrasada.

Revisemos si las conclusiones van en la direccion correcta

```python
corr, p_value = pearsonr(df['Lagging_Current_Reactive_Power_kVarh'], df['CO2_tCO2'])
print(f"Correlación de Pearson: {corr}")
print(f"Valor p: {p_value}")

if p_value < 0.05:
    print("La correlación es estadísticamente significativa.")
else:
    print("No hay evidencia suficiente para concluir que la correlación es significativa.")
```

```python
corr, p_value = pearsonr(df['Usage_kWh'], df['CO2_tCO2'])
print(f"Correlación de Pearson: {corr}")
print(f"Valor p: {p_value}")

if p_value < 0.05:
    print("La correlación es estadísticamente significativa.")
else:
    print("No hay evidencia suficiente para concluir que la correlación es significativa.")
```
